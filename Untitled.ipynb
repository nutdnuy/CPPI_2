```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Parameters
initial_portfolio_value = 100000  # Initial portfolio value in USD
multiplier = 2  # Multiplier for CPPI
floor_percentage = 0.7  # 30% max drawdown implies a 70% floor
risk_free_rate = 0.02  # Annual risk-free rate

# Fetch BTC-USD data
btc_data = yf.download('BTC-USD', start='2018-01-01', end='2024-12-31')
btc_data['Returns'] = btc_data['Adj Close'].pct_change().fillna(0)

# CPPI Strategy Implementation
def backtest_cppi(prices, multiplier, floor_percentage, initial_value):
    n = len(prices)
    account_value = np.zeros(n)
    risky_allocation = np.zeros(n)
    cushion = np.zeros(n)
    floor = initial_value * floor_percentage
    
    account_value[0] = initial_value
    risky_allocation[0] = 0  # Start with no exposure to BTC
    
    for t in range(1, n):
        cushion[t] = max(0, account_value[t-1] - floor)
        risky_allocation[t] = min(multiplier * cushion[t], account_value[t-1])
        safe_allocation = account_value[t-1] - risky_allocation[t]
        risky_return = prices[t] / prices[t-1] - 1
        account_value[t] = (risky_allocation[t] * (1 + risky_return) +
                            safe_allocation * (1 + risk_free_rate / 252))
        floor = max(floor, floor_percentage * account_value[t])
    
    return account_value

# Calculate NAVs
cppi_nav = backtest_cppi(btc_data['Adj Close'].values, multiplier, floor_percentage, initial_portfolio_value)
buy_hold_nav = initial_portfolio_value * (1 + btc_data['Returns']).cumprod()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(btc_data.index, cppi_nav, label='CPPI Strategy NAV', color='blue')
plt.plot(btc_data.index, buy_hold_nav, label='Buy and Hold NAV', color='green')
plt.title('CPPI Strategy vs. Buy and Hold Strategy')
plt.xlabel('Date')
plt.ylabel('Portfolio Value (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate BTC price data (mock data)
np.random.seed(42)
dates = pd.date_range(start='2018-01-01', end='2024-12-31', freq='D')
btc_prices = 10000 * np.cumprod(1 + np.random.normal(0, 0.02, len(dates)))

# Create a DataFrame to store the simulated BTC price data
btc_data = pd.DataFrame({
    'Date': dates,
    'Adj Close': btc_prices
})

# Calculate daily returns for BTC
btc_data['Returns'] = btc_data['Adj Close'].pct_change().fillna(0)

# Parameters for CPPI
initial_portfolio_value = 100000  # Initial portfolio value in USD
multiplier = 2  # Multiplier (m) for CPPI
floor_percentage = 0.7  # 30% max drawdown = 70% floor
risk_free_rate = 0.02  # Annual risk-free rate

# Backtest CPPI with yearly reset
def backtest_cppi_yearly(prices, multiplier, floor_percentage, initial_portfolio_value):
    prices['Year'] = prices['Date'].dt.year
    yearly_results = []

    for year, year_data in prices.groupby('Year'):
        n = len(year_data)
        account_value = np.zeros(n)  # Total portfolio value
        risky_allocation = np.zeros(n)  # Amount allocated to BTC
        cushion = np.zeros(n)  # Cushion to measure distance from floor
        floor = initial_portfolio_value * floor_percentage  # Dynamic floor

        account_value[0] = initial_portfolio_value
        risky_allocation[0] = 0  # Assume no initial exposure to BTC

        for t in range(1, n):
            # Update the cushion
            cushion[t] = max(0, account_value[t-1] - floor)

            # Update risky asset allocation
            risky_allocation[t] = multiplier * cushion[t]

            # Ensure allocation does not exceed account value
            risky_allocation[t] = min(risky_allocation[t], account_value[t-1])

            # Calculate the allocation to the risk-free asset (e.g., cash)
            safe_allocation = account_value[t-1] - risky_allocation[t]

            # Update account value based on BTC price change
            risky_return = year_data['Adj Close'].iloc[t] / year_data['Adj Close'].iloc[t-1] - 1
            account_value[t] = risky_allocation[t] * (1 + risky_return) + safe_allocation * (1 + risk_free_rate / 252)

            # Update the floor (to limit max drawdown to 30% per year)
            floor = max(floor, floor_percentage * account_value[t])

        yearly_results.append(pd.DataFrame({
            'Date': year_data['Date'].values,
            'Account Value': account_value,
            'Risky Allocation': risky_allocation,
            'Cushion': cushion,
            'Floor': floor
        }))

    return pd.concat(yearly_results)

# Run CPPI with yearly reset
backtest_results_cppi = backtest_cppi_yearly(btc_data, multiplier, floor_percentage, initial_portfolio_value)

# Calculate Buy-and-Hold NAV with yearly reset
btc_data['Year'] = btc_data['Date'].dt.year
buy_and_hold_results = []

for year, year_data in btc_data.groupby('Year'):
    initial_btc_price = year_data['Adj Close'].iloc[0]
    buy_hold_nav = initial_portfolio_value * year_data['Adj Close'] / initial_btc_price
    buy_and_hold_results.append(pd.DataFrame({
        'Date': year_data['Date'].values,
        'Buy and Hold NAV': buy_hold_nav
    }))

buy_and_hold_results = pd.concat(buy_and_hold_results)

# Merge CPPI results with BTC Buy and Hold NAV
combined_results = backtest_results_cppi[['Date', 'Account Value']].rename(columns={'Account Value': 'CPPI NAV'})
combined_results['Buy and Hold NAV'] = buy_and_hold_results.set_index('Date')['Buy and Hold NAV'].values
combined_results.set_index('Date', inplace=True)

# Plot CPPI vs Buy-and-Hold NAV
plt.figure(figsize=(12, 6))
plt.plot(combined_results.index, combined_results['CPPI NAV'], label='CPPI NAV', color='blue')
plt.plot(combined_results.index, combined_results['Buy and Hold NAV'], label='Buy and Hold NAV', color='green')
plt.title('CPPI Strategy vs Buy-and-Hold Strategy (Yearly Reset)')
plt.xlabel('Date')
plt.ylabel('Portfolio Value (USD)')
plt.legend()
plt.show()
```
